```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv")
```

```python
df = df.drop("name", axis=1)
```

```python
df.head()
```

```python
df.origin.value_counts()
```

```python
def origin(x):
    if x == "usa":
        return 1
    else:
        return 0
```

```python
df["origin"] = df.origin.apply(origin)
```

```python
df.head()
```

```python
X = df.drop("mpg", axis=1)
```

```python
y = df.mpg
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.pipeline import Pipeline
```

```python
norm = MinMaxScaler()
norm.fit(X)
X_sc = norm.transform(X)
```

```python
RandomizedSearchCV??
```

```python
params = {
    "max_depth": [3, 5, 7, 9],
    "n_estimators": [100, 200, 300]
}

# Initialize the RandomForestRegressor
rf = RandomForestRegressor(n_jobs=-1)

# Set up RandomizedSearchCV
grid = RandomizedSearchCV(
    estimator=rf,
    param_distributions=params,
    n_iter=10,  # Number of parameter settings that are sampled
    cv=5,  # Number of folds for cross-validation
    verbose=2,
    n_jobs=-1,
    random_state=42
)

grid.fit(X_sc, y)
```

```python
pipe = Pipeline([
    ("norm", MinMaxScaler()),
    ("rf", RandomForestRegressor(n_jobs=-1))
])

# Define the parameter grid
params = {
    "rf__max_depth": [3, 5, 7, 9],
    "rf__n_estimators": [100, 200, 300]
}

# Set up RandomizedSearchCV
grid = RandomizedSearchCV(
    estimator=pipe,
    param_distributions=params,
    n_iter=10,  # Number of parameter settings that are sampled
    cv=5,  # Number of folds for cross-validation
    verbose=2,
    n_jobs=-1,
    random_state=42
)

# Fit the grid to the data
grid.fit(X, y)
```

```python
grid.best_estimator_.score
```

```python
grid.best_score_
```

```python
import pickle
with open("model.plk", "wb") as f:
    pickle.dump(grid.best_estimator_, f)
```

```python
with open("model.plk", "rb") as f:
    saved_model = pickle.load(f)
```
